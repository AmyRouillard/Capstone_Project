## Import libraries

```python
import os

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import image as mpimg

import IPython.display as ipd
from IPython.core.display import display
```

## Find all folders in train folder

```python
folders = []
for fname in os.listdir('./train/'):
  if os.path.isdir('./train/'+fname):
    folders.append(fname)
```

```python
folders
```

## Show all files associated with a given subfolder (i.e. catagory)

```python
i_folder = 8

for f in os.listdir('./train/'+folders[i_folder]):
  f_wav = f.split('.')
  f_wav = './splits/'+f_wav[0]+'.wav'
  display(ipd.Audio("./"+f_wav, autoplay=True))
  image = mpimg.imread('./train/'+folders[i_folder]+'/'+f)
  plt.imshow(image)
  plt.show()
```
